```python
def fusionner_dicts(dict1, dict2):
  fusion = {}
  for element in dict1:
    fusion[element] = 1
  for element in dict2:
    if element in dict1:
      fusion[element] += 1
    else:
      fusion [element] = 1
  return fusion
dict1={'Ahmed','Brahim','Fati','Imane' }
dict2={'Mohammed','Brahim','Hamid','Imane'}
fusionner_dicts(dict1, dict2)
```

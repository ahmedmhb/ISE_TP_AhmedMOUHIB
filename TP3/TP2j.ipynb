```python
class chien:
  def __init__(self,nom,race,age):
    self.nom=nom
    self.race=race
    self.age=age
  def aboyer(self):
    return "Woof!"
chien1=chien("rex","pitbull",3)
chien1.aboyer()
```

```python
class voiture:
  def __init__(self,marque,modele,annee):
    self.marque=marque
    self.modele=modele
    self.annee=annee
  def afficherInfo(self):
    print(f'{self.marque, self.modele, self.annee}')
dadCar = voiture("opel","corsa",2011)
dadCar.afficherInfo()
```

```python
class compteBancaire:
  def __init__(self,titulaire,solde):
    self.titulaire= titulaire
    self.solde = solde
  def deposer(self,montant):
    self.solde += montant
  def retirer(self,montant):
    if self.solde > 0 and self.solde >= montant:
      self.solde -= montant
    else:
      return "solde insuffisant"
  def afficher_solde(self):
    return self.solde
client1=compteBancaire("Ahmed MHB",3000)
client1.retirer(4000)
client1.retirer(1000)
client1.deposer(2000)
client1.afficher_solde()
```

```python
class person:
  def __init__(self,nom,prenom,age):
    self.nom=nom
    self.prenom=prenom
    self.age=age
  def se_presenter(self):
    print(f"je suis {self.nom} {self.prenom}, j'ai {self.age} ans")
class etudiant(person):
  def __init__(self,nom,prenom,age,matricule):
    super().__init__(nom,prenom,age)
    self.matricule=matricule
  def etudier(self):
    print( self.matricule)
etudiant1= etudiant("Mouhib","Ahmed",21,"K13333776")
etudiant1.se_presenter()
etudiant1.etudier()
```

```python
class animal:
  def __init__(self):
    pass
  def faire_de_bruit(self,type):
    print(f"{type} fais du bruit")
class chien(animal):
  def __init__(self):
    super().__init__()
class chat(animal):
  def __init__(self):
    super().__init__()
cat=chat()
cat.faire_de_bruit('cat')
```

```python
class rectangle:
  def __init__(self,longueur,hauteur):
    self.longueur=longueur
    self.hauteur=hauteur
  def calculer_surface(self):
    print(self.longueur*self.hauteur)
  def calculer_perimetre(self):
    print(2*(self.longueur+self.hauteur))
rectangle1=rectangle(4,5)
rectangle1.calculer_surface()
rectangle1.calculer_perimetre()
```

```python
class livre:
  def __init__(self,titre,auteur,annee_publication):
    self.titre=titre
    self.auteur=auteur
    self.annee_publication=annee_publication
class bibliotheque(livre):
  def __init__(self, titre, auteur, annee_publication):
    super().__init__(titre, auteur, annee_publication)
  livres=[]
  def ajouter_livre(self):
    self.livres.append(self.titre)
  def afficher_livres(self):
    print(self.livres)
livre1=bibliotheque("la boite a merveille","sidi mohammed",1900)
livre1.ajouter_livre()
livre1.afficher_livres()
```
